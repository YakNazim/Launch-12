# GPS

Sort through the raw GPS data 50ms at a time. Then look at `soft-correlator`'s output of doppler, code phase, and SNR.

This jumps right through the data gap at t+4.55 seconds

```python
import json
import glob

files = glob.glob('./sats/*')

# pare down to valid sats that are in the sky
sats = {}
for sat in [1,4,7,11,13,15,17,24,28,30]:
    sats["%2d"%sat] = {'doppler': [0]*len(files), 'code': [0]*len(files), 'snr': [0]*len(files)}
    
for i, fname in enumerate(sorted(files)):
    #print fname, "~~~~~~~~~~~~~~"
    with open(fname, 'r') as f_in:
        try:
            data = json.loads(f_in.read())
        except:
            continue
        for sat, cor in data.iteritems():
            if sat not in sats:
                continue
            sats[sat]['doppler'][i] = cor['doppler']
            sats[sat]['code'][i] = cor['phase']
            sats[sat]['snr'][i] = cor['snr']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

time = [(i*50e-3)-32.041913222 for i in range(len(files))]

fig, ax1 = plt.subplots(figsize=(16,6))
plt.title(r"GPS Sat Code Phase Drift")
plt.xlabel(r"Mission Elapsed Time [s]")
plt.ylabel(r"Code Phase [chips]")

for sat in sorted(sats):
    cor = sats[sat]
    times = []
    codes = []
    for i, c in enumerate(cor['code']):
        if c > 0:
            times.append(time[i])
            codes.append(c)
    if int(sat) < 16:
        ax1.plot(times, codes, '-o', alpha=0.65, lw=0.12, ms=3, mew=0,label=sat)
    else:
        ax1.plot(times, codes, '-v', alpha=0.65, lw=0.12, ms=5, mew=0,label=sat)

ax1.set_xlim([-30,35])
ax1.set_ylim([0,1024])
ax1.legend(loc=3, title="PRN:")
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(16,6))
plt.title(r"GPS Sat Doppler Shift")
plt.xlabel(r"Mission Elapsed Time [s]")
plt.ylabel(r"Doppler [Hz]")

for sat in sorted(sats):
    cor = sats[sat]
    times = []
    dops = []
    for i, d in enumerate(cor['doppler']):
        if d != 0:
            times.append(time[i])
            dops.append(d)
    if int(sat) < 16:
        ax1.plot(times, dops, '-o', alpha=0.65, lw=0.12, ms=3, mew=0,label=sat)
    else:
        ax1.plot(times, dops, '-v', alpha=0.65, lw=0.12, ms=5, mew=0,label=sat)

ax1.set_xlim([-30,35])
ax1.set_ylim([-5000, 4000])
ax1.legend(loc=3, title="PRN:")
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(16,6))
plt.title(r"GPS Sat Signal to Noise")
plt.xlabel(r"Mission Elapsed Time [s]")
plt.ylabel(r"Doppler [Hz]")

for sat in sorted(sats):
    cor = sats[sat]
    times = []
    snrs = []
    for i, s in enumerate(cor['snr']):
        if s != 0:
            times.append(time[i])
            snrs.append(s)
    ax1.plot(times, snrs, '-', alpha=0.4, label=sat)

ax1.set_xlim([-30,35])
#ax1.set_ylim([-6000, 6000])
#ax1.legend(loc=2)
plt.show()
```
