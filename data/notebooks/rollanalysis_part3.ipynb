```python
import warnings
warnings.filterwarnings('ignore')
from thLib import quat
import numpy
%matplotlib inline
import matplotlib.pylab as plt
from scipy.signal import savgol_filter, correlate
from scipy.integrate import simps

# Graphing helper function
def setup_graph(title='',x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0],
                            fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

# grab data
columns = numpy.loadtxt("../LAUNCH/ADIS.csv", delimiter=',', unpack=True)

# time
time = columns[1]

# umblicial disconnect event
t_0 = 117853569585227

# element wise subtraction
time = numpy.subtract(time, t_0)

# convert from ns to s
time = numpy.divide(time, 1e9)

# IMU measurements
gyro_x = columns[3]
gyro_y = columns[4]
gyro_z = columns[5]
acc_x = columns[6]
acc_y = columns[7]
acc_z = columns[8]
mag_x = columns[9]
mag_y = columns[10]
mag_z = columns[11]
```

### Calculating revolutions with gyroscope measurements

In this section I revisit some past analysis to use for comparison later. Quick caveat: for gyroscope measurements on launch-12 the x-axis is the axis of interest.

In the code below I do the following:

1. Eliminated the samples I don't care about which are the ones before takeoff and after apogee
2. Filtered the gyro measurements with a Savitzky–Golay filter
    * This filter simply smooths the data out to increase the signal-to-noise ratio without distorting the signal. This is important because more noise means less accurate calculations with the data.
3. Integrated the filtered gyro measurements
    * A gyro measures angular velocity but I need angular position so I integrate the gyro measurements, which are already given in $\frac{degrees}{s}$, to obtain that.
4. Calculated the number of revolutions
    * The maximum angular position divided by 360 degrees should be the number of revolutions during the flight.

```python
# eliminate samples before takeoff and after apogee
time = time[3650:28125]
gyro_x = gyro_x[3650:28125]
gyro_y = gyro_y[3650:28125]
gyro_z = gyro_z[3650:28125]

# filter gyro readings
gyro_x = savgol_filter(gyro_x, 201, 2)
gyro_y = savgol_filter(gyro_y, 201, 2)
gyro_z = savgol_filter(gyro_z, 201, 2)

# integrate to get angular position
ang = []
for k in range(len(gyro_x)):
    if(k > 0):
        ang.append(simps(gyro_x[:k], time[:k]))
ang = numpy.array(ang)

setup_graph('Gyroscope measurements from Launch-12', r'Time $s$', r'Angular velocity $\frac{deg}{s}$', (9,6.5))
plt.plot(time, gyro_x)
plt.show()

setup_graph('Derived angular position from Launch-12', r'Time $s$', 'Angular position (deg)', (9,6.5))
plt.plot(time[1:], ang)
plt.show()

gyro_result = ang.max()/360.0

print 'Total revolutions: %3.2f' % gyro_result
```

### Magnetometer measurements

The magnetomer measurements are something that I have not explored in previous analysis as I mentioned earlier so in this section I did the following:

1. Eliminated the samples I don't care about which are the ones before takeoff and after apogee
2. Filtered the magnetometer measurements with a Savitzky–Golay filter
    * This filter simply smooths the data out to increase the signal-to-noise ratio without distorting the signal. This is important because more noise means less accurate calculations with the data.
3. Plotted the magnetometer measurements

Since the magnetometers are mostly looking at the Earth's relativly fixed magnetic field, one should expect to see it change as the rocket spins. For magnetometer measurements on launch-12, the y and z axes are the ones of interest.

```python
# eliminate samples before takeoff and after apogee
mag_x = mag_x[3650:28125]
mag_y = mag_y[3650:28125]
mag_z = mag_z[3650:28125]

# filter mag readings
mag_x = savgol_filter(mag_x, 201, 2)
mag_y = savgol_filter(mag_y, 201, 2)
mag_z = savgol_filter(mag_z, 201, 2)

# plot readings
setup_graph('Magnetometer readings from Launch-12', r'Time $s$', 'Measurement of magnetic field (gauss)', (9,6.5))
plt.plot(time, mag_y)
plt.plot(time, mag_z)
plt.legend(['y-axis', 'z-axis'], loc='best')
plt.show()
```

#### Phase shift

```python
fft_y = numpy.fft.fft(mag_y)
fft_z = numpy.fft.fft(mag_z)
ang_y = numpy.degrees(numpy.angle(fft_y))
ang_z = numpy.degrees(numpy.angle(fft_z))

phase_shift = []
for i, j in zip(ang_y, ang_z):
    phase_shift.append(numpy.abs(i-j))
phase_shift = numpy.array(phase_shift)

print 'Average phase shift between signals: %3.2f degrees' % phase_shift.mean()
```

### Calculating revolutions with gyroscope measurements

Taking the data from the magnetometer in the y and z axes and treating it as a vector allows us to observe how that vector changes as the rocket spins. We can get that vector to rotate about the origin by setting the mean to 0. I will refer to this vector as the ***yz vector*** later on.

```python
mag_y = mag_y-mag_y.mean()
mag_z = mag_z-mag_z.mean()

setup_graph('Magnetometer y-z vector', 'y-component (gauss)', 'z-component (gauss)', (9,6.5))
plt.plot(mag_y, mag_z)
plt.show()
```

Using the yz vector, we can compute the roll angle

```python
# use numpy to build [y,z] vector from the two arrays
mag = numpy.dstack((mag_y, mag_z))[0]

# reference angle, should be a mean, but good enough.
begin = mag[0]

# storage
angles = []     # roll angle
rollup = 0      # every time we make one revolution, add 360 to this
last_angle = 0  # for checking if we've wrapped back to -pi

for v1 in mag:
    
    # compute angle between now and the reference angle
    s = numpy.cross(v1, begin)
    c = numpy.dot(v1, begin)
    angle = numpy.arctan2(s, c)
    
    # if we just wrapped around, add a revolution
    if last_angle > 2 and angle < 0:
        rollup += 360
        
    # store
    angles.append(numpy.degrees(angle)+rollup)
    last_angle = angle
    
mag_result = rollup/360

pd = 100* numpy.abs(gyro_result - mag_result)/gyro_result

print 'Total revolutions %3.2f' % mag_result
print "Percent difference from Gyro result %3.2f%%" % pd
```

The result above shows that the magnetometer can provide just as good an estimate as the gyro can.

### Orientation Estimates (UNFINISHED)

Utilizing the magnetometer measurements also allows us to use a quaternion based kalman filter to produce orientation estimates . The orientation estimates are in the form of a quaternion which can be translated in to a rotation matrix. The dot product of that rotation matrix with a reference axis (i.e. x-axis [1,0,0]) produces the change of that reference axis in a global reference frame.

#### Setup

```python
from mpl_toolkits.mplot3d import Axes3D
from thLib import imus

# 3D Graphing helper function
def setup_graph_3D(title='',x_label='', y_label='', z_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0],
                            fig_size[1])
    ax = fig.add_subplot(111, projection='3d')
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_zlabel(z_label)
    return ax

# axes
x = [0,0,-1]
y = [1,0,0]
z = [0,1,0]

gyro_x = columns[3]
gyro_y = columns[4]
gyro_z = columns[5]
acc_x = columns[6]
acc_y = columns[7]
acc_z = columns[8]
mag_x = columns[9]
mag_y = columns[10]
mag_z = columns[11]

gyro_x = gyro_x[3650:28125]
gyro_y = gyro_y[3650:28125]
gyro_z = gyro_z[3650:28125]
acc_x = acc_x[3650:28125]
acc_y = acc_y[3650:28125]
acc_z = acc_z[3650:28125]
mag_x = mag_x[3650:28125]
mag_y = mag_y[3650:28125]
mag_z = mag_z[3650:28125]

gyro_x = savgol_filter(gyro_x, 201, 2)
gyro_y = savgol_filter(gyro_y, 201, 2)
gyro_z = savgol_filter(gyro_z, 201, 2)
acc_x = savgol_filter(acc_x, 201, 2)
acc_y = savgol_filter(acc_y, 201, 2)
acc_z = savgol_filter(acc_z, 201, 2)
mag_x = savgol_filter(mag_x, 201, 2)
mag_y = savgol_filter(mag_y, 201, 2)
mag_z = savgol_filter(mag_z, 201, 2)

# create vectors for IMU data
gyro = numpy.array(zip(gyro_x, gyro_y, gyro_z))
acc = numpy.array(zip(acc_x, acc_y, acc_z))
mag = numpy.array(zip(mag_x, mag_y, mag_z))

# normalize gyro vectors
ugyro = []
for row in gyro:
    tude = numpy.sqrt(row[0]**2 + row[1]**2 + row[2]**2)
    ugyro.append(numpy.divide(row, tude))
ugyro = numpy.array(ugyro)

# normalize acceleration vectors
uacc = []
for row in acc:
    tude = numpy.sqrt(row[0]**2 + row[1]**2 + row[2]**2)
    uacc.append(numpy.divide(row, tude))
uacc = numpy.array(uacc)

data = {}
data['acc'] = uacc
data['omega'] = ugyro
data['mag'] = mag
data['rate'] = 819.2

ADIS = imus.IMU(inData=data)

# Quaternion to Rotation Matrix
def QuaternionToMatrix(qa):
    q = qa.copy()
    q = numpy.reshape(q, (4,))
    R = numpy.zeros((3,3))
    R[0,0] = 1-2*q[2]**2 - 2*q[3]**2
    R[1,0] = 2*q[1]*q[2] + 2*q[3]*q[0]
    R[2,0] = 2*q[1]*q[3] - 2*q[2]*q[0]
    R[0,1] = 2*q[1]*q[2] - 2*q[3]*q[0]
    R[1,1] = 1 - 2*q[1]**2 - 2*q[3]**2
    R[2,1] = 2*q[2]*q[3] + 2*q[1]*q[0]
    R[0,2] = 2*q[1]*q[3] + 2*q[2]*q[0]
    R[1,2] = 2*q[2]*q[3] - 2*q[1]*q[0]
    R[2,2] = 1 - 2*q[1]**2 - 2*q[2]**2
    return R
```

```python
initialOrientation = numpy.array([x, y, z])
ADIS.calc_orientation(initialOrientation, type='Kalman')
quaternion = numpy.nan_to_num(ADIS.quat)
#quaternion = ADIS.quat

#quaternion = quat.vel2quat(ugyro, [0,0,0], 819.2, 'bf')
#quaternion = quat.vel2quat(gyro, [0,0,0], 819.2, 'bf')

x_ref = []
for element in quaternion:
    x_ref.append(numpy.dot(QuaternionToMatrix(element), x))
x_ref = numpy.array(x_ref)

y_ref = []
for element in quaternion:
    y_ref.append(numpy.dot(QuaternionToMatrix(element), y))
y_ref = numpy.array(y_ref)

z_ref = []
for element in quaternion:
    z_ref.append(numpy.dot(QuaternionToMatrix(element), z))
z_ref = numpy.array(z_ref)

ax = setup_graph_3D('Change of X-axis in reference frame', 'x-component', 'y-component', 'z-component', (9,6.5))
ax.plot(x_ref[:,0], x_ref[:,1], x_ref[:,2])
plt.show()

ax = setup_graph_3D('Change of Y-axis in reference frame', 'x-component', 'y-component', 'z-component', (9,6.5))
ax.plot(y_ref[:,0], y_ref[:,1], y_ref[:,2])
plt.show()

ax = setup_graph_3D('Change of Z-axis in reference frame', 'x-component', 'y-component', 'z-component', (9,6.5))
ax.plot(z_ref[:,0], z_ref[:,1], z_ref[:,2])
plt.show()
```

```python
setup_graph('Change of X-axis in reference frame', 'Sample', 'Result', (9,6.5))
plt.plot(x_ref)
plt.show()

setup_graph('Change of Y-axis in reference frame', 'Sample', 'Result', (9,6.5))
plt.plot(y_ref)
plt.show()

setup_graph('Change of Z-axis in reference frame', 'Sample', 'Result', (9,6.5))
plt.plot(z_ref)
plt.show()
```
