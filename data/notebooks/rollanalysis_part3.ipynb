## Analysis

### Setup

In this section I prepare everything that I need for analysis.

In the code below, I do the following:

1. Load in packages that I plan to use for analysis
2. Define a helper function for making nice plots quickly
3. Load the Launch-12 data

```python
from thLib import quat
import numpy
%matplotlib inline
import matplotlib.pylab as plt
from scipy.signal import savgol_filter, correlate
from scipy.integrate import simps

# Graphing helper function
def setup_graph(title='',x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0],
                            fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

# grab data
columns = numpy.loadtxt("../LAUNCH/ADIS.csv", delimiter=',', unpack=True)
time = columns[1]
gyro_x = columns[3]
gyro_y = columns[4]
gyro_z = columns[5]
acc_x = columns[6]
acc_y = columns[7]
acc_z = columns[8]
mag_x = columns[9]
mag_y = columns[10]
mag_z = columns[11]

# umblicial disconnect event
t_0 = 117853569585227
time = numpy.subtract(time, t_0)

# convert from ns to s
time = numpy.divide(time, 1e9)
```

### Calculating revolutions with gyroscope measurements

In this section I revisit some past analysis to use for comparison later.

In the code below I do the following:

1. Eliminate the samples which are not important for roll control which are the ones before takeoff and after apogee
2. Filter the x-axis gyro measurements with a Savitzky–Golay filter
    * This filter simply smooths the data out to increase the signal-to-noise ratio without distorting the signal. This is important because more noise means less accurate calculations with the data.
    * Because of the orientation of the IMU on LV2.3, the x-axis is the axis of interest for roll.
3. Integrate the filtered gyro measurements
    * A gyro measures angular velocity but I need angular position so I integrate the gyro measurements, which are already given in deg/s, to obtain that.
4. Calculated the number of revolutions
    * The maximum angular position divided by 360 degrees should be the number of revolutions during the flight.

```python
# eliminate samples before takeoff and after apogee
time = time[3650:28125]
gyro_x = gyro_x[3650:28125]
gyro_y = gyro_y[3650:28125]
gyro_z = gyro_z[3650:28125]

# index of apogee event, known from past analysis
apogee_index = 24472

# filter gyro readings
gyro_x = savgol_filter(gyro_x, 201, 2)
gyro_y = savgol_filter(gyro_y, 201, 2)
gyro_z = savgol_filter(gyro_z, 201, 2)

# integrate to get angular position
ang = []
for k in range(len(gyro_x)):
    if(k > 0):
        ang.append(simps(gyro_x[:k], time[:k]))
ang = numpy.array(ang)

# markers for annotation
takeoff_gyro = (time[0], gyro_x[0])
takeoff_ang = (time[0], ang[0])
apogee_gyro = (time[apogee_index], gyro_x[apogee_index])
apogee_ang = (time[apogee_index], ang[apogee_index])

setup_graph('Gyroscope measurements from Launch-12', r'Time $s$', r'Angular velocity $\frac{deg}{s}$', (9,6.5))
plt.plot(time, gyro_x)
plt.annotate('Lift-off',
             takeoff_gyro,
             xycoords='data',
             xytext=(10,-30),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.annotate('Apogee',
             apogee_gyro,
             xycoords='data',
             xytext=(10,-20),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.show()

setup_graph('Derived angular position from Launch-12', r'Time $s$', 'Angular position (deg)', (9,6.5))
plt.plot(time[1:], ang)
plt.annotate('Lift-off',
             takeoff_ang,
             xycoords='data',
             xytext=(10,-20),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.annotate('Apogee',
             apogee_ang,
             xycoords='data',
             xytext=(10,-20),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.show()

gyro_result = ang.max()/360.0

print 'Total revolutions: %3.2f' % gyro_result
```

### Magnetometer measurements

The magnetomer measurements are something that have not yet been explored in Launch-12 analysis. In this section, I do the following:

1. Eliminate the samples which are not important for roll control which are the ones before takeoff and after apogee
2. Filter the magnetometer measurements with a Savitzky–Golay filter
    * This filter simply smooths the data out to increase the signal-to-noise ratio without distorting the signal. This is important because more noise means less accurate calculations with the data.
3. Plot the y and z axis magnetometer measurements
    * Since the magnetometers are mostly looking at the Earth's relativly fixed magnetic field, one should expect to see it change as the rocket spins.
    * Because of the orientation of the IMU on LV2.3, the y and z axes are the ones of interest for roll.

```python
# eliminate samples before takeoff and after apogee
mag_x = mag_x[3650:28125]
mag_y = mag_y[3650:28125]
mag_z = mag_z[3650:28125]

# filter mag readings
mag_x = savgol_filter(mag_x, 201, 2)
mag_y = savgol_filter(mag_y, 201, 2)
mag_z = savgol_filter(mag_z, 201, 2)

# plot readings
setup_graph('Magnetometer readings from Launch-12', r'Time $s$', 'Measurement of magnetic field (gauss)', (9,6.5))
plt.plot(time, mag_y)
plt.plot(time, mag_z)
plt.legend(['y-axis', 'z-axis'], loc='best')
plt.show()
```

Looking at these plots, one might notice how nicely shifted these waveforms are. As an added bonus, I attempted to calculate the phase shift below.

#### Phase shift

```python
fft_y = numpy.fft.fft(mag_y)
fft_z = numpy.fft.fft(mag_z)
ang_y = numpy.degrees(numpy.angle(fft_y))
ang_z = numpy.degrees(numpy.angle(fft_z))

phase_shift = []
for i, j in zip(ang_y, ang_z):
    phase_shift.append(numpy.abs(i-j))
phase_shift = numpy.array(phase_shift)

print 'Average phase shift between signals: %3.2f degrees' % (phase_shift.mean()/2)
```

It is interesting to see that the phase shift is almost 90 degrees which is very similar to the [phase shift of a sine wave to a cosine wave](https://en.wikipedia.org/wiki/Sine_wave#Occurrences). This means we should get a very nice looking circle when we treat the y and z axis measurements as a vector.

### Calculating revolutions with magnetometer measurements

Taking the data from the magnetometer in the y and z axes and treating it as a vector allows us to observe how that vector changes as the rocket spins. I will refer to this vector as the ***yz vector*** later on. We can get that vector to rotate about the origin by removing the mean. As noted earlier, the nice 90 degree phase shift between these components should give us a nice circle. In the code below, I remove the mean and plot the yz vector.

```python
mag_y = mag_y-mag_y.mean()
mag_z = mag_z-mag_z.mean()

takeoff_mag = (mag_y[0], mag_z[0])
apogee_mag = (mag_y[apogee_index], mag_z[apogee_index])

setup_graph('Magnetometer y-z vector', 'y-component (gauss)', 'z-component (gauss)', (9,6.5))
plt.plot(mag_y, mag_z)
plt.annotate('Lift-off',
             takeoff_mag,
             xycoords='data',
             xytext=(-70,20),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.annotate('Apogee',
             apogee_mag,
             xycoords='data',
             xytext=(0,-30),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.show()
```

Using the yz vector, we can compute the roll angle with the function defined below. We will utilize this trick again later on which is why I have declared this function. After I calculate the revolutions with this function, I compare it to the result from our gyroscope. The result should be very close if not exactly the same.

```python
def calculate_revolutions(vector):
    # reference angle, should be a mean, but good enough.
    begin = vector[0]

    # storage
    angles = []          # roll angle
    rollup = float(0)    # every time we make one revolution, add 360 to this
    last_angle = 0       # for checking if we've wrapped back to -pi

    for v1 in vector:
    
        # compute angle between now and the reference angle
        s = numpy.cross(v1, begin)
        c = numpy.dot(v1, begin)
        angle = numpy.arctan2(s, c)
    
        # if we just wrapped around, add a revolution
        if last_angle > 2 and angle < 0:
            rollup += 360
        
        # store
        angles.append(numpy.degrees(angle)+rollup)
        last_angle = angle
    
    return rollup/360

# use numpy to build [y,z] vector from the two arrays
yz_vector = numpy.dstack((mag_y, mag_z))[0]

# use our handy-dandy function
mag_result = calculate_revolutions(yz_vector)

# calculate percent difference
pd = 100* numpy.abs(gyro_result - mag_result)/gyro_result

# print results
print 'Total revolutions %3.2f' % mag_result
print "Percent difference from Gyro result %3.2f%%" % pd
```

This result shows that the magnetometer and gyroscope result are very close as expected.

### Orientation Estimates

We can use all of our IMU data to produce orientation estimates via a [quaternion](https://en.wikipedia.org/wiki/Quaternions_and_spatial_rotation) based [Kalman Filter](https://en.wikipedia.org/wiki/Kalman_filter).

Kalman filters and quaternions are mathematically sophisticated yet extremely powerful tools for estimating orientation. I will neglect much of that sophistication here which will probably abhor my peers but oh well.

In a nut shell, a Kalman filter is a recursive algorithm that uses a series of measurements observed over time along with the model of a system (e.g. process and measurement models) to predict state variables called estimates. Upon producing the estimates, they are updated using a weighted average, with more weight being given to estimates with higher certainty. In the code below, I make use of a quaternion based kalman filter that is a part of the [thLib package for python](http://work.thaslwanter.at/thLib/html/index.html). The kalman filter is designed based on "Design, Implementation, and Experimental Results of a Quaternion-Based Kalman Filter for Human Body Motion Tracking” Yun, X. and Bachman, E.R., IEEE TRANSACTIONS ON ROBOTICS, VOL. 22, 1216-1227 (2006). This isn't actually ideal for our rocket and in fact in the paper it even says "The algorithm is based on the assumption that the rigid body to which the sensor is attached is stationary or is slow moving, and is thus not applicable to highly dynamic tracking applications". However, with some slight changes to the source code which I will not cover, I was able to make it work 'good enough'. If you'd like to know more about Kalman filters, I highly recommend watching the [fantastic set of videos on State Space Tracking](https://www.youtube.com/user/mcnames/videos) posted by [Dr. James McNames](http://web.cecs.pdx.edu/~mcnames/) on youtube.

Now in school you were told that vectors and scalars don't mix, which is a lie, because quaternions. A quaternion $q$ can be used to represent a spatial rotation and contains a vector part and a scalar part. More specifically, $q \in \mathbb{R^4}$ where q = $(q_0, q_1, q_2, q_3)$, $q_0$ is the scalar part and $q_1, q_2, q_3$ the vector part.

If we let a rotation through an angle of $\theta$ around an axis defined by a unit vector be

${\displaystyle {\vec {u}}=(u_{x},u_{y},u_{z})=u_{x}\mathbf {i} +u_{y}\mathbf {j} +u_{z}\mathbf {k}}$

Using an extension of Euler's formula, this can be represented by the following quaternion:

$q = e^{{\frac {\theta }{2}}{(u_{x}\mathbf {i} +u_{y}\mathbf {j} +u_{z}\mathbf {k} )}}=\cos {\frac {\theta }{2}}+(u_{x}\mathbf {i} +u_{y}\mathbf {j} +u_{z}\mathbf {k} )\sin {\frac {\theta }{2}} = (q_0, q_1, q_2, q_3)$

This quaternion can be translated into a rotation matrix as such

${\displaystyle \mathbf {R} ={\begin{bmatrix}1-2q_{2}^{2}-2q_{3}^{2}&2(q_{1}q_{2}-q_{3}q_{0})&2(q_{1}q_{3}+q_{2}q_{0})\\2(q_{1}q_{2}+q_{3}q_{0})&1-2q_{1}^{2}-2q_{3}^{2}&2(q_{2}q_{3}-q_{1}q_{0})\\2(q_{1}q_{3}-q_{2}q_{0})&2(q_{2}q_{3}+q_{1}q_{0})&1-2q_{1}^{2}-2q_{2}^{2}\end{bmatrix}}}$

The dot product of each row of this matrix with a vector that represents an axis (for example, the x-axis [1,0,0]) will produce the x, y, and z components of how that axis has changed in a reference frame.

In summary, this is exactly what I will do below

1. Re-load all the IMU data, re-trim it and filter it just like I did before
2. Create vectors and more important unit vectors with the IMU data
3. Initialize an IMU object (thLib) with these vectors
4. Provide an initial orientation to the kalman filter and request estimates
5. Translate the quaternions (orientation estimates) into rotation matrices and calculate the change of axes of orientation for LV2.3 during Launch-12

#### Steps 1-3

```python
from mpl_toolkits.mplot3d import Axes3D
from thLib import imus

# 3D Graphing helper function
def setup_graph_3D(title='',x_label='', y_label='', z_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0],
                            fig_size[1])
    ax = fig.add_subplot(111, projection='3d')
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_zlabel(z_label)
    return ax

# axes of orientation for the IMU on LV2.3
x = [0,0,-1]
y = [0,1,0]
z = [1,0,0]

"""
reload the data
trim it to the samples that are important
and filter it
"""
gyro_x = columns[3]
gyro_y = columns[4]
gyro_z = columns[5]
acc_x = columns[6]
acc_y = columns[7]
acc_z = columns[8]
mag_x = columns[9]
mag_y = columns[10]
mag_z = columns[11]
gyro_x = gyro_x[3650:28125]
gyro_y = gyro_y[3650:28125]
gyro_z = gyro_z[3650:28125]
acc_x = acc_x[3650:28125]
acc_y = acc_y[3650:28125]
acc_z = acc_z[3650:28125]
mag_x = mag_x[3650:28125]
mag_y = mag_y[3650:28125]
mag_z = mag_z[3650:28125]
gyro_x = savgol_filter(gyro_x, 201, 2)
gyro_y = savgol_filter(gyro_y, 201, 2)
gyro_z = savgol_filter(gyro_z, 201, 2)
acc_x = savgol_filter(acc_x, 201, 2)
acc_y = savgol_filter(acc_y, 201, 2)
acc_z = savgol_filter(acc_z, 201, 2)
mag_x = savgol_filter(mag_x, 201, 2)
mag_y = savgol_filter(mag_y, 201, 2)
mag_z = savgol_filter(mag_z, 201, 2)

# create vectors for IMU data
gyro = numpy.array(zip(gyro_x, gyro_y, gyro_z))
acc = numpy.array(zip(acc_x, acc_y, acc_z))
mag = numpy.array(zip(mag_x, mag_y, mag_z))

def create_unit_vectors(vector):
    result = []
    for row in vector:
        magnitude = numpy.sqrt(row[0]**2 + row[1]**2 + row[2]**2)
        result.append(numpy.divide(row, magnitude))
    result = numpy.array(result)
    return result

# create unit vectors
ugyro = create_unit_vectors(gyro)
uacc = create_unit_vectors(acc)
umag = create_unit_vectors(mag)

# initialize IMU object using thLib
data = {}
data['acc'] = uacc
data['omega'] = ugyro
data['mag'] = umag
data['rate'] = 819.2
ADIS = imus.IMU(inData=data)
```

#### Steps 4 & 5

```python
initialOrientation = numpy.array([x, y, z])
ADIS.calc_orientation(initialOrientation, type='Kalman')
quaternion = numpy.nan_to_num(ADIS.quat)

# Quaternion to Rotation Matrix
def QuaternionToMatrix(qa):
    q = qa.copy()
    q = numpy.reshape(q, (4,))
    R = numpy.zeros((3,3))
    R[0,0] = 1-2*q[2]**2 - 2*q[3]**2
    R[1,0] = 2*q[1]*q[2] + 2*q[3]*q[0]
    R[2,0] = 2*q[1]*q[3] - 2*q[2]*q[0]
    R[0,1] = 2*q[1]*q[2] - 2*q[3]*q[0]
    R[1,1] = 1 - 2*q[1]**2 - 2*q[3]**2
    R[2,1] = 2*q[2]*q[3] + 2*q[1]*q[0]
    R[0,2] = 2*q[1]*q[3] + 2*q[2]*q[0]
    R[1,2] = 2*q[2]*q[3] - 2*q[1]*q[0]
    R[2,2] = 1 - 2*q[1]**2 - 2*q[2]**2
    return R

# calculate change of axes
x_ref = []
for element in quaternion:
    x_ref.append(numpy.dot(QuaternionToMatrix(element), x))
x_ref = numpy.array(x_ref)

y_ref = []
for element in quaternion:
    y_ref.append(numpy.dot(QuaternionToMatrix(element), y))
y_ref = numpy.array(y_ref)

z_ref = []
for element in quaternion:
    z_ref.append(numpy.dot(QuaternionToMatrix(element), z))
z_ref = numpy.array(z_ref)

# plot results
ax = setup_graph_3D('Change of X-axis in reference frame', 'x-component', 'y-component', 'z-component', (9,6.5))
ax.plot(x_ref[:,0], x_ref[:,1], x_ref[:,2])
plt.show()
ax = setup_graph_3D('Change of Y-axis in reference frame', 'x-component', 'y-component', 'z-component', (9,6.5))
ax.plot(y_ref[:,0], y_ref[:,1], y_ref[:,2])
plt.show()
ax = setup_graph_3D('Change of Z-axis in reference frame', 'x-component', 'y-component', 'z-component', (9,6.5))
ax.plot(z_ref[:,0], z_ref[:,1], z_ref[:,2])
plt.show()
```

These results can be projected onto a 2D plane in a similar manner to the magnetometer seen earlier. The same trick for calculating revolutions can then be used. What's interesting about the visualizations above is how different they appear. It is easy to see that they will each provide a different result. But which one is right? Probably the y-axis since this is the one pointing out horizontally from the rocket (i.e. [0,1,0]). However, I will average the results instead of making that assumption to see if the average is the same as the results obtained earlier.

```python
xref0 = x_ref[:,0]-x_ref[:,0].mean()
xref1 = x_ref[:,1]-x_ref[:,1].mean()

takeoff_xref = (xref0[0], xref1[0])
apogee_xref = (xref0[apogee_index], xref1[apogee_index])

setup_graph('Change of X-axis in reference frame (2D)', 'x-component', 'y-component', (9,6.5))
plt.plot(xref0, xref1)
plt.annotate('Lift-off',
             takeoff_xref,
             xycoords='data',
             xytext=(0,-20),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.annotate('Apogee',
             apogee_xref,
             xycoords='data',
             xytext=(-10,-30),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.show()
```

```python
yref0 = y_ref[:,0]-y_ref[:,0].mean()
yref1 = y_ref[:,1]-y_ref[:,1].mean()

takeoff_yref = (yref0[0], yref1[0])
apogee_yref = (yref0[apogee_index], yref1[apogee_index])

setup_graph('Change of Y-axis in reference frame (2D)', 'x-component', 'y-component', (9,6.5))
plt.plot(yref0, yref1)
plt.annotate('Lift-off',
             takeoff_yref,
             xycoords='data',
             xytext=(-70,10),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.annotate('Apogee',
             apogee_yref,
             xycoords='data',
             xytext=(20,-30),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.show()
```

```python
zref0 = z_ref[:,0]-z_ref[:,0].mean()
zref1 = z_ref[:,1]-z_ref[:,1].mean()

takeoff_zref = (zref0[0], zref1[0])
apogee_zref = (zref0[apogee_index], zref1[apogee_index])

setup_graph('Change of Z-axis in reference frame (2D)', 'x-component', 'y-component', (9,6.5))
plt.plot(zref0, zref1)
plt.annotate('Lift-off',
             takeoff_zref,
             xycoords='data',
             xytext=(-10,-70),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.annotate('Apogee',
             apogee_zref,
             xycoords='data',
             xytext=(20,-30),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.show()
```

```python
xref_xy_vector = numpy.dstack((xref0, xref1))[0]
yref_xy_vector = numpy.dstack((yref0, yref1))[0]
zref_xy_vector = numpy.dstack((zref0, zref1))[0]

xref_result = calculate_revolutions(xref_xy_vector)
yref_result = calculate_revolutions(yref_xy_vector)
zref_result = calculate_revolutions(zref_xy_vector)

print 'Calculated revolutions with x-axis xy vector %s' % xref_result
print 'Calculated revolutions with y-axis xy vector %s' % yref_result
print 'Calculated revolutions with z-axis xy vector %s\n' % zref_result

print 'Average of results %s' % ((xref_result + yref_result + zref_result)/3)
```

This is the expected result and this analysis can now be concluded.
