## Setup

```python
import numpy
%matplotlib inline
import matplotlib.pylab as plt
from scipy.constants import K2C

# Graphing helper function
def setup_graph(title='',x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0],
                            fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

# load data
columns = numpy.loadtxt("../LAUNCH/RNHH.csv", delimiter=',', unpack=True)
SEQN = columns[0]
time = columns[1]
Temperature = columns[2]
TS1Temperature = columns[3]
TS2Temperature = columns[4]
TempRange = columns[5]
Voltage = columns[6]
Current = columns[7]
AverageCurrent = columns[8]
CellVoltage1 = columns[9]
CellVoltage2 = columns[10]
CellVoltage3 = columns[11]
CellVoltage4 = columns[12]
PackVoltage = columns[13]
AverageVoltage = columns[14]

# umblicial disconnect event
t_0 = 117853569585227
time = numpy.subtract(time, t_0)

# convert from ns to s
time = numpy.divide(time, 1e9)
```

## Temperature

```python
Temperature = K2C(Temperature)
TS1Temperature = K2C(TS1Temperature)
TS2Temperature = K2C(TS2Temperature)

setup_graph('Temperature', 'Time (s)', 'Temp (C)', (9,6.5))
plt.plot(time, Temperature)
plt.plot(time, TS1Temperature)
plt.plot(time, TS2Temperature)
plt.legend(['Temperature', 'TS1Temperature', 'TS2Temperature'], loc='best')
plt.show()
```

## Power Consumption

```python
max_current_draw = (time[numpy.argmin(Current)], Current.min())
apogee = (time[33], Current[33])

setup_graph('Current', 'Time (s)', 'Current (Amps)', (9,6.5))
plt.plot(time, Current)
plt.plot(time, AverageCurrent)
plt.annotate('Max current draw',
             max_current_draw,
             xycoords='data',
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.annotate('Apogee',
             apogee,
             xycoords='data',
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.legend(['Current', 'Average'], loc='best')
plt.show()
```

```python
print 'Max Current draw (Amps): %3.2f' % Current.min()
print 'Minimum occurred at t=%3.2f s\n' % time[numpy.argmin(Current)]
```

The sign of the current is negative because we are drawing current from the battery during flight.

```python
# calculate power
Power = []
for x,y in zip(Current,Voltage):
    Power.append(abs(x)*y)
Power = numpy.array(Power)
```

```python
max_power = (time[numpy.argmax(Power)], Power.max())
apogee = (time[33], Power[33])

setup_graph('Power', 'Time (s)', 'Watts', (9,6.5))
plt.plot(time, Power)
plt.annotate('Max Power Usage',
             max_power,
             xycoords='data',
             xytext=(20,-10),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.annotate('Apogee',
             apogee,
             xycoords='data',
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))
plt.show()
```
