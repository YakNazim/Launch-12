## Setup

```python
import numpy
%matplotlib inline
import matplotlib.pylab as plt

# Graphing helper function
def setup_graph(title='',x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0],
                            fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

# load data
columns = numpy.loadtxt("../LAUNCH/FCFH.csv", delimiter=',', unpack=True)
SEQN = columns[0]
time = columns[1]
CPU_User = columns[2]
CPU_System = columns[3]
CPU_Nice = columns[4]
CPU_IOWait = columns[5]
CPU_IRQ = columns[6]
CPU_SoftIRQ = columns[7]
RAM_Used = columns[8]
RAM_Buffer = columns[9]
RAM_Cached = columns[10]
PID = columns[11]
Disk_Used = columns[12]
Disk_Read = columns[13]
Disk_Write = columns[14]
IO_lo_Bytes_Sent = columns[15]
IO_lo_Bytes_Recv = columns[16]
IO_lo_Packets_Sent = columns[17]
IO_lo_Packets_Recv = columns[18]
IO_eth0_Bytes_Sent = columns[19]
IO_eth0_Bytes_Recv = columns[20]
IO_eth0_Packets_Sent = columns[21]
IO_eth0_Packets_Recv = columns[22]
IO_wlan0_Bytes_Sent = columns[23]
IO_wlan0_Bytes_Recv = columns[24]
IO_wlan0_Packets_Sent = columns[25]
IO_wlan0_Packets_Recv = columns[26]
Core_Temp = columns[27]


# umblicial disconnect event
t_0 = 117853569585227
time = numpy.subtract(time, t_0)

# convert from ns to s
time = numpy.divide(time, 1e9)
```

## CPU Usage

```python
setup_graph('CPU Usage vs Time', 'Time (s)', 'Usage (%)', (9, 6.5))
plt.plot(time, CPU_User)
plt.plot(time, CPU_System)
plt.plot(time, CPU_Nice)
plt.plot(time, CPU_IOWait)
plt.plot(time, CPU_IRQ)
plt.plot(time, CPU_SoftIRQ)
plt.legend(['User', 'System', 'Nice', 'IOWait', 'IRQ', 'SoftIRQ'], loc='best')
plt.show()
```

```python
# calculate total cpu usage
CPU_Total = []
for i,j,k,l,m,n in zip(CPU_User, CPU_System, CPU_Nice, CPU_IOWait, CPU_IRQ, CPU_SoftIRQ):
    CPU_Total.append(i+j+k+l+m+n)
CPU_Total = numpy.array(CPU_Total)

setup_graph('Total CPU Usage vs Time', 'Time (s)', 'Usage (%)', (9, 6.5))
plt.plot(time, CPU_Total)
plt.show()
```

## RAM usage

```python
# convert to gigabytes
RAM_Used = numpy.divide(RAM_Used, 1e9)
RAM_Buffer = numpy.divide(RAM_Buffer, 1e9)
RAM_Cached = numpy.divide(RAM_Cached, 1e9)

setup_graph('RAM vs Time', 'Time (s)', 'Gigabytes', (9, 6.5))
plt.plot(time, RAM_Used)
plt.plot(time, RAM_Buffer)
plt.plot(time, RAM_Cached)
plt.legend(['Used', 'Buffer', 'Cached'], loc='best')
plt.show()
```

## Disk usage

```python
# convert to gigabytes
Disk_Used = numpy.divide(Disk_Used, 1e9)

setup_graph('Disk Used vs Time', 'Time (s)', 'Gigabytes', (9, 6.5))
plt.plot(time, Disk_Used)
plt.show()
```

### Disk IO

```python
# convert to megabytes
Disk_Read = numpy.divide(Disk_Read, 1e6)
Disk_Write = numpy.divide(Disk_Write, 1e6)

setup_graph('Disk Read/Write vs Time', 'Time (s)', 'Megabytes', (9, 6.5))
plt.plot(time, Disk_Read)
plt.plot(time, Disk_Write)
plt.legend(['Read', 'Write', 'Total'], loc='best')
plt.show()

print 'Average Read Throughput: %3.2f MB/s (std: %3.2f MB/s)' % (Disk_Read.mean(), Disk_Read.std())
print 'Average Write Throughput: %3.2f MB/s (std: %3.2f MB/s)' % (Disk_Write.mean(), Disk_Write.std())
```

```python
total = numpy.add(Disk_Read, Disk_Write)
setup_graph('Throughput vs Time', 'Time (s)', 'Megabytes', (9, 6.5))
plt.plot(time, total)
plt.show()

print 'Average throughput: %3.2f MB/s (std: %3.2f MB/s)' % (total.mean(), total.std())
```
